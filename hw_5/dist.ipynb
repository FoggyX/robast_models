```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform, binom, poisson, norm, expon
```

Равномерное распределение

```python
x = np.linspace(-1, 2, 100)
plt.plot(x, uniform.pdf(x))
plt.title('Плотность равномерного распределения') 
plt.show()
```

```python
x = np.linspace(-1, 2, 100)
plt.plot(x, uniform.cdf(x))
plt.title('Функция распределения') 
plt.show()
```

Биномиальное распределение

```python
x = np.linspace(-1, 8, 10)
plt.scatter(x, binom(n=10, p=0.5).pmf(x))
plt.title('Плотность биномиального распределения') 
plt.show()
```

```python
x = np.linspace(-1, 11, 100)
plt.plot(x, binom(n=10, p=0.5).cdf(x))
plt.title('Функция распределения') 
plt.show()
```

Распределение Пуассона

```python
x = np.linspace(-2, 8, 11)
plt.scatter(x, poisson(0.5).pmf(x))
plt.title('Плотность распределения Пуассона') 
plt.show()
```

```python
x = np.linspace(-1, 6, 100)
plt.plot(x, poisson(0.5).cdf(x))
plt.title('Функция распределения') 
plt.show()
```

Нормальное распределение

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, norm.pdf(x))
plt.title('Плотность нормального распределения') 
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, norm.cdf(x))
plt.title('Функция распределения') 
plt.show()
```

Экспоненциальное распределение

```python
x = np.linspace(-2, 10, 1000)
plt.plot(x, expon.pdf(x))
plt.title('Плотность экспоненциального распределения') 
plt.show()
```

```python
x = np.linspace(-2, 10, 1000)
plt.plot(x, expon.cdf(x))
plt.title('Функция распределения') 
plt.show()
```
